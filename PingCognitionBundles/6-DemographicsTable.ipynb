Creating a demographic table 

The following variables (stratified by sex) will be incorporated in the final table:

1. N (%)
2. Age, months (std)
3. Handedness
4. Race/Ethnicity, count (%)
5. Highest parental education, count (%)
6. Familial Income (USD$), count (%)
7. ADHD, count (%)
8. Cognitive scores, mean (std)

```python
# imports
import os

import numpy as np
import pandas as pd
```

```python
# Load up PING data
ping_data = pd.read_excel((f"{ping_bundles_dir}/ping_data_filtered_for_bundles.xlsx"))
```

```python
# Etchnicity demographics 
ping_male = ping_data[ping_data.Sex==1]
ping_female = ping_data[ping_data.Sex==2]
ping_male_ethn = ping_male.Ethnicity.value_counts()
ping_female_ethn = ping_female.Ethnicity.value_counts()

# Parental education demographics
ping_male_edu = ping_male.edu_groups.value_counts()
ping_female_edu = ping_female.edu_groups.value_counts()

# Income demographics
ping_male_inc = ping_male.income_groups.value_counts()
ping_female_inc = ping_female.income_groups.value_counts()

# Handedness demographics
ping_male_hand = ping_male.Handedness.value_counts()
ping_female_hand = ping_female.Handedness.value_counts()
```

```python
# Fetching the mean value and appending them to a list
male_desc = [
    [ping_male.count().iloc[0], np.round(ping_male.count().iloc[0] * 100 / ping_data.shape[0], 2)],
    [f"{ping_male.AgeMonths.mean()/12:.2f} ({ping_male.AgeMonths.min()/12:.2f}–{ping_male.AgeMonths.max()/12:.2f})"],

    [ping_male_hand.loc[2], np.round(ping_male_hand.loc[2] * 100 / ping_data.shape[0], 2)],
    [ping_male_hand.loc[1], np.round(ping_male_hand.loc[1] * 100 / ping_data.shape[0], 2)],
    [ping_male_hand.loc[3], np.round(ping_male_hand.loc[3] * 100 / ping_data.shape[0], 2)],
    
    [ping_male_ethn.loc[1], np.round(ping_male_ethn.loc[1] * 100 / ping_data.shape[0], 2)],
    [ping_male_ethn.loc[2], np.round(ping_male_ethn.loc[2] * 100 / ping_data.shape[0], 2)],
    [ping_male_ethn.loc[3], np.round(ping_male_ethn.loc[3] * 100 / ping_data.shape[0], 2)],
    [ping_male_ethn.loc[4], np.round(ping_male_ethn.loc[4] * 100 / ping_data.shape[0], 2)],
    [ping_male_ethn.loc[5], np.round(ping_male_ethn.loc[5] * 100 / ping_data.shape[0], 2)],
    [ping_male_ethn.loc[6], np.round(ping_male_ethn.loc[6] * 100 / ping_data.shape[0], 2)],

    [ping_male_edu.loc[1], np.round(ping_male_edu.loc[1] * 100 / ping_data.shape[0], 2)],
    [ping_male_edu.loc[2], np.round(ping_male_edu.loc[2] * 100 / ping_data.shape[0], 2)],
    [ping_male_edu.loc[3], np.round(ping_male_edu.loc[3] * 100 / ping_data.shape[0], 2)],
    [ping_male_edu.loc[4], np.round(ping_male_edu.loc[4] * 100 / ping_data.shape[0], 2)],
    [ping_male_edu.loc[5], np.round(ping_male_edu.loc[5] * 100 / ping_data.shape[0], 2)],

    [ping_male_inc.loc[1], np.round(ping_male_inc.loc[1] * 100 / ping_data.shape[0], 2)],
    [ping_male_inc.loc[2], np.round(ping_male_inc.loc[2] * 100 / ping_data.shape[0], 2)],
    [ping_male_inc.loc[3], np.round(ping_male_inc.loc[3] * 100 / ping_data.shape[0], 2)],

    [ping_male.ADHD.value_counts().iloc[1], np.round(ping_male.ADHD.value_counts().iloc[1] * 100 / ping_data.shape[0], 2)],

    [np.round(ping_male.VA.mean(), 2), np.round(ping_male.VA.std(), 2)],
    [np.round(ping_male.EFPS.mean(), 2), np.round(ping_male.EFPS.std(), 2)],
    [np.round(ping_male.MEM.mean(), 2), np.round(ping_male.MEM.std(), 2)],
]

female_desc = [
    [ping_female.count().iloc[0], np.round(ping_female.count().iloc[0] * 100 / ping_data.shape[0], 2)],
    [f"{ping_female.AgeMonths.mean()/12:.2f} ({ping_female.AgeMonths.min()/12:.2f}–{ping_female.AgeMonths.max()/12:.2f})"],

    [ping_female_hand.loc[2], np.round(ping_female_hand.loc[2] * 100 / ping_data.shape[0], 2)],
    [ping_female_hand.loc[1], np.round(ping_female_hand.loc[1] * 100 / ping_data.shape[0], 2)],
    [ping_female_hand.loc[3], np.round(ping_female_hand.loc[3] * 100 / ping_data.shape[0], 2)],
    
    [ping_female_ethn.loc[1], np.round(ping_female_ethn.loc[1] * 100 / ping_data.shape[0], 2)],
    [ping_female_ethn.loc[2], np.round(ping_female_ethn.loc[2] * 100 / ping_data.shape[0], 2)],
    [ping_female_ethn.loc[3], np.round(ping_female_ethn.loc[3] * 100 / ping_data.shape[0], 2)],
    [ping_female_ethn.loc[4], np.round(ping_female_ethn.loc[4] * 100 / ping_data.shape[0], 2)],
    [ping_female_ethn.loc[5], np.round(ping_female_ethn.loc[5] * 100 / ping_data.shape[0], 2)],
    "-",

    [ping_female_edu.loc[1], np.round(ping_female_edu.loc[1] * 100 / ping_data.shape[0], 2)],
    [ping_female_edu.loc[2], np.round(ping_female_edu.loc[2] * 100 / ping_data.shape[0], 2)],
    [ping_female_edu.loc[3], np.round(ping_female_edu.loc[3] * 100 / ping_data.shape[0], 2)],
    [ping_female_edu.loc[4], np.round(ping_female_edu.loc[4] * 100 / ping_data.shape[0], 2)],
    [ping_female_edu.loc[5], np.round(ping_female_edu.loc[5] * 100 / ping_data.shape[0], 2)],

    [ping_female_inc.loc[1], np.round(ping_female_inc.loc[1] * 100 / ping_data.shape[0], 2)],
    [ping_female_inc.loc[2], np.round(ping_female_inc.loc[2] * 100 / ping_data.shape[0], 2)],
    [ping_female_inc.loc[3], np.round(ping_female_inc.loc[3] * 100 / ping_data.shape[0], 2)],

    [ping_female.ADHD.value_counts().iloc[1], np.round(ping_female.ADHD.value_counts().iloc[1] * 100 / ping_data.shape[0], 2)],

    [np.round(ping_female.VA.mean(), 2), np.round(ping_female.VA.std(), 2)],
    [np.round(ping_female.EFPS.mean(), 2), np.round(ping_female.EFPS.std(), 2)],
    [np.round(ping_female.MEM.mean(), 2), np.round(ping_female.MEM.std(), 2)],
]
```

```python
# Creating the demographics table
demo_table = pd.DataFrame({
    "ping_male" : male_desc,
    "ping_female" : female_desc
})
demo_table.index = [
    "N (%)", "Âge, moyenne (étendue)", "Droitier", "Gaucher", "Ambidextre", "Blanc", "Noir ou Africain Américain", "Hispanique ou Latino", "Asiatique", "Multiethnique", "Autre Ethnicité",
    "Aucun diplôme d'études secondaires", "Diplôme d'études secondaires ou équivalent", "DEC, DEP ou équivalent", "Baccalauréat", "Diplôme d'études supérieures",
    "<50 000 $", "50 000 - 100 000$", ">100 000$",
    "ADHD, nombre (%)", "VA", "EFPS", "MEM"
]

demo_table.to_excel(f"{output_dir}/demographics_table.xlsx", index=True, header=True)
```
